```python
theta = np.linspace(0,1,100) * np.pi * 2
X = theta
angular_speed = np.linspace(-8,8,100,endpoint=True)
Y = angular_speed
```

```python
cmp = plt.colormaps["Blues"]
level = [np.linspace(-2,2,10), np.linspace(-2,2,10),np.linspace(-2,2,10)]
def plot(Z,title,i):
    fig1, ax2 = plt.subplots(layout='constrained')
    origin = 'lower'

    CS = ax2.contourf(X, Y, Z,10, cmap=cmp, origin=origin,levels=level[i])
    ax2.set_title(title)
    ax2.set_xlabel('position (angle)')
    ax2.set_ylabel('speed')

    # Make a colorbar for the ContourSet returned by the contourf call.
    cbar = fig1.colorbar(CS)
    cbar.ax.set_ylabel('entropy')
# Add the contour line levels to the colorbar
```
