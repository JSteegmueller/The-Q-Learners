```python
import numpy as np
import laserhockey.hockey_env as h_env
import gymnasium as gym
from gymnasium import spaces
import os
import pickle
import torch
from SACAgent import SAC
from param_save import save_param, load_param
import matplotlib.pyplot as plt
```

```python
def play(env,agent,n_episodes =100,gamma=1,max_Step=2000,LH=False, render=False):
    rewards = []
    timestep = 0

    #player1 = agent
    opponent = h_env.BasicOpponent(weak=False)
    #max_Step=2000
    for ep in range(1, n_episodes+1):
        #print(ep)
        
        #env.reset()
        ob, _info = env.reset()
        obs_agent2 = env.obs_agent_two()

        total_reward=0
        
        for t in range(max_Step):
            
            timestep += 1
            done = False
            #a,_ =  player1.act(ob,True)
            #a2 = player2.act(obs_agent2)
            a,_ = agent.act(ob,True)

            ob2 = env.obs_agent_two()
            a2 = opponent.act(ob2,True)

            (ob_new, reward, done, trunc, _info) = env.step(np.hstack([a,a2]))
            total_reward +=  gamma ** t * reward
            ob=ob_new
            if done or trunc: 
                break  
            if render:
                clear_output(wait=True)
                plt.imshow(env.render())
                plt.show()
            if LH:
                env.render(mode="human")
        #print("episode ",ep,total_reward)      
        rewards.append(total_reward)
    if render or LH:
        plt.close()
        env.close()
    
    print(f'Mean reward: {np.mean(rewards)}')
    return rewards
```

```python
env1 = h_env.HockeyEnv_BasicOpponent(mode=h_env.HockeyEnv.TRAIN_DEFENSE, weak_opponent=False)
env2 = h_env.HockeyEnv(mode=h_env.HockeyEnv.TRAIN_DEFENSE)
```
